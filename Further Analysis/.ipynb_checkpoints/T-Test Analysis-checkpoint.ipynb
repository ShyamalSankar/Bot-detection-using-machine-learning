# Import Packages

```python
# Import packages
import random
import math
import warnings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.utils import resample
from scipy.stats import beta, f_oneway, kruskal, ttest_ind, chi2_contingency

pd.set_option("display.max_columns", None)
warnings.filterwarnings("ignore")
```

```python
metadata_dataset = pd.read_csv("metadata_dataset.csv")
metadata_dataset
```

```python
features_dataset = metadata_dataset[["protected", "verified", "followers_count", "following_count", "tweet_count", "listed_count", "isBot"]]
features_dataset
```

# T-test

```python
bot_followers_count = list(features_dataset[features_dataset['isBot'] == 1]['followers_count'])
human_followers_count = list(features_dataset[features_dataset['isBot'] == 0]['followers_count'])
#test for equal variance


ttest_ind(bot_followers_count, human_followers_count, equal_var=False)
```

```python
bot_following_count = list(features_dataset[features_dataset['isBot'] == 1]['following_count'])
human_following_count = list(features_dataset[features_dataset['isBot'] == 0]['following_count'])

ttest_ind(bot_following_count, human_following_count, equal_var = False)
```

```python
bot_tweet_count = list(features_dataset[features_dataset['isBot'] == 1]['tweet_count'])
human_tweet_count = list(features_dataset[features_dataset['isBot'] == 0]['tweet_count'])

ttest_ind(bot_tweet_count, human_tweet_count, equal_var = False)
```

# Chi-sq test

```python
bot_human_contigency = pd.crosstab(features_dataset['isBot'], features_dataset['protected'])
```

```python
#The first is chisq statistic, second is p val third is dof of test fourth is expected values
chi2_contingency(bot_human_contigency)
```

```python
bot_human_contingency_verified = pd.crosstab(features_dataset['isBot'], features_dataset['verified']) 
```

```python
bot_human_contingency_verified
```

```python
#The first is chisq statistic, second is p val third is dof of test fourth is expected values
chi2_contingency(bot_human_contingency_verified)
```
